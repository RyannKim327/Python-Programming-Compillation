```python
# Imports
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
# Get data

data = pd.read_csv("data/music.csv")
data
```

```python
# Setups
# drop use to remove the column that is not requied to predict
drop = data.drop(columns=['genre']) # X
need = data["genre"] # y

decide = DecisionTreeClassifier()
decide.fit(drop.values, need.values)

result = decide.predict([
	[21, 1],
	[22, 0]
])

print(f"Male [21]: {result[0]}\nFemale [22]: {result[1]}")
```
