```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
import warnings

warnings.filterwarnings("ignore")
```

```python
# Import

data = pd.read_csv("data/Demographic-Data.csv")
```

```python
data.columns = ['Country_Name', 'Country_Code', 'Birth_rate', 'Internet', 'Income']
data.columns
```

```python
data.Country_Name = data.Country_Name.astype("category")
data.Country_Code = data.Country_Code.astype("category")
data.Income = data.Income.astype("category")
data.info()
data
```

# Does the birth rate affects with the Income?

```python
sb.boxplot(data=data, x=data.Income, y=data.Birth_rate)
plt.gcf().set_size_inches(15, 8)
```

# Does the income affects with the use of internet?

```python
sb.boxplot(data=data, x=data.Income, y=data.Internet)
plt.xticks(rotation=-35, horizontalalignment="left")
plt.gcf().set_size_inches(15, 5)
```

# What country having the most populated?

```python
plt.bar(data.Country_Code, data.Birth_rate)
plt.xticks(rotation=-35, horizontalalignment="left")
plt.gcf().set_size_inches(100, 20)
plt.grid()
plt.show()
```
